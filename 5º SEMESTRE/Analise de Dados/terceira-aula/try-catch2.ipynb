```python
import pandas as pd
import mysql.connector

```

```python
DB_HOST="localhost"
DB_USER="adados"
DB_PASS="12345"
```

```python
try:
    conn = mysql.connector.connect(host=DB_HOST, 
                                port="3306", 
                                user=DB_USER, 
                                password=DB_PASS, 
                                autocommit=False)
    if not conn.is_connected():
        raise mysql.connector.Error("Falha crítica: O banco de dados recusou a sessão.")
    else:
        print("Conexão com Sucesso")
        cursor = conn.cursor()
except mysql.connector.Error as e:
# O erro é acessado diretamente pelo caminho do módulo
    print(f"Erro detectado pelo Driver: {e.msg}")
    print(f"Código numérico do MySQL: {e.errno}")

except Exception as e:
    # Segurança para erros que não são do MySQL (ex: erro de memória)
    print(f"Erro de sistema: {e}")

```

```python
# Ao pressionar Ctrl + C, a exceção KeyboardInterrupt será capturada pelo bloco BaseException.
# no jupiter não funciona corretamente
try:
    print("Processando... (pressione Ctrl+C para interromper)")
    
    while True:
        print("Executando...")
        
except Exception:
    print("Capturado por Exception")

except BaseException as erro:
    print(f"Capturado por BaseException: {type(erro).__name__}")
```

```python
# A chamada sys.exit() gera SystemExit, que é capturada apenas por BaseException.
import sys

try:
    print("Encerrando o programa...")
    sys.exit("Saída solicitada")

except Exception:
    print("Capturado por Exception")

except BaseException as erro:
    print(f"Capturado por BaseException: {type(erro).__name__} - {erro}")
```

```python
# função para o generador
def meu_generator():
    try:
        while True:
            yield "Gerando valor..."
    except GeneratorExit:
        print("Generator encerrado corretamente.")
        raise  # recomendável relançar a exceção
```

```python
# Criando o generator
gen = meu_generator()
```

```python

print(next(gen))  # Produz o primeiro valor
```

```python
# Encerrando explicitamente
gen.close()
```

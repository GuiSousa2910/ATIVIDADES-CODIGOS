```python
import pandas as pd
import unicodedata
import re
```

```python
def remover_acentos(texto):
    if isinstance(texto, str):
        texto = unicodedata.normalize('NFD', texto)
        texto = texto.encode('ascii', 'ignore').decode('utf-8')
    return texto

def limpar_texto(texto):
    if isinstance(texto, str):
        texto = remover_acentos(texto)
        texto = texto.upper()
        texto = texto.strip()
        texto = re.sub(r'\s+', ' ', texto)
        texto = texto.replace('\n','').replace('\r','').replace('\t','')
    return texto
```

```python
df = pd.read_csv("detalhe_votacao.csv", sep=',', encoding='utf-8')
```

```python
# Remover duplicatas
df = df.drop_duplicates()

# Tratar valores nulos
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna("NAO INFORMADO")
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(limpar_texto)

# Converter datas se existir coluna de data
if 'DATA_VOTACAO' in df.columns:
    df['DATA_VOTACAO'] = pd.to_datetime(df['DATA_VOTACAO'], errors='coerce')
    df['ANO_VOTACAO'] = df['DATA_VOTACAO'].dt.year
    df['MES_VOTACAO'] = df['DATA_VOTACAO'].dt.month
```

```python
df.to_excel("detalhe_votacao_tratado.xlsx", index=False)

print("Arquivo tratado gerado com sucesso.")
```

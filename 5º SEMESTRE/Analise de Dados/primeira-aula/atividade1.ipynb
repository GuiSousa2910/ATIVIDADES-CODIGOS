```python
#1.1) Cadastro básico - Crie variáveis para nome e idade. Mostre o tipo de cada variável e calcule o ano aproximado de nascimento considerando 2025.
nome = str(input('Qual seu nome?'))
print('Tipo da variavel nome: ' + str(type(nome)))
idade = int(input('Qual sua idade?'))
print('Tipo da variavel idade: ' + str(type(idade)))
data_nasc = 2025 - idade
print(data_nasc)
```

```python
#1.2) Altura - Crie uma variável de altura em metros, mostre o tipo e converta para centímetros.
altura = float(input('Digite sua altura em metro: '))
altura_cm = altura * 100
print(altura_cm)
```

```python
#1.3) Documento - Crie um documento como texto, gere uma versão numérica e some 100 ao valor.
n = int(input('Insira o numero:'))
print(n + 100)
```

```python
#1.4) Quantidade de caixas – Entre com a quantidade de caixas e calcule o total de itens (12 por caixa).
caixa = int(input('Quantidade de caixas: '))
itens_totais = caixa * 12

print('Itens totais: ', itens_totais)
```

```python
#1.5) Preço com desconto - Entre o preço do produto, calcule 10% de desconto e mostre o valor final.
valor_total = float(input('Valor: '))
valor_desconto = valor_total * 0.9
print(valor_desconto)
```

```python
#1.6) Soma simples – Entre com dois valores e calcule a soma.
n1 = float(input('Numero 1:'))
n2 = float(input('Numero 2:'))
soma = n1 + n2
print(soma)
```

```python
#1.7) Média de notas - Crie duas variáveis notas como texto, converta para número e calcule a média.
n1 = float(input('Numero 1:'))
n2 = float(input('Numero 2:'))
media = (n1 + n2) / 2
print(media)
```

```python
#1.8) Valor total da compra – Crie uma variável para o valor unitário fixo, atribua um valor, calcule o total da compra com a quantidade informada pelo usuário
valor_fixo = 10

quantidade_produtos = int(input('Quantos itens voce pegou:'))
print('Valor total:', valor_fixo * quantidade_produtos)
```

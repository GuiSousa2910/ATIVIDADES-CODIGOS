```python
#3.17) Soma acumulada - Mostre a soma acumulada dos valores da segunda coluna.
matriz = [[1, 10], [2, 20], [3, 30]]
acumulado = 0
for linha in matriz:
    acumulado += linha[1]
    print('Acumulado:', acumulado)
```

```python
#3.18) Par ou impar - Informe se os valores da segunda coluna sao pares ou impares.
matriz = [[1, 10], [2, 21], [3, 32]]
for linha in matriz:
    valor = linha[1]
    if valor % 2 == 0:
        print(valor, 'par')
    else:
        print(valor, 'impar')
```

```python
#3.19) Interrupcao - Interrompa o processamento ao encontrar valor negativo.
matriz = [[1, 10], [2, 5], [3, -1], [4, 20]]
for linha in matriz:
    if linha[1] < 0:
        print('Valor negativo encontrado. Parando.')
        break
    print(linha)
```

```python
#3.20) Controle por condicao - Percorra a matriz enquanto os valores da segunda coluna forem maiores que zero.
matriz = [[1, 10], [2, 5], [3, 1], [4, 0], [5, 7]]
i = 0
while i < len(matriz) and matriz[i][1] > 0:
    print(matriz[i])
    i += 1
```
